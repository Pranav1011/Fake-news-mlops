```python
import sagemaker
from sagemaker.sklearn.estimator import SKLearn
import boto3

s3 = s3 = boto3.client('s3')
# Specify S3 paths for input data and output model artifacts
train_data_s3_path = "s3://fake-news-pranav/data/processed"
output_s3_path = "s3://fake-news-pranav/models/"
```

```python
# Define the SKLearn estimator
sklearn_estimator = SKLearn(
    entry_point="./scripts/train.py",  # Path to train.py script
    role=sagemaker.get_execution_role(),
    instance_type="ml.m5.large",
    instance_count=1,
    framework_version="1.2-1",  # Match scikit-learn version
    py_version="py3",
    output_path=output_s3_path,
)

# Launch the training job
sklearn_estimator.fit({"train": train_data_s3_path})
```
